```python
# import the libraries 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn import preprocessing, metrics
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
sns.set(style="white", color_codes=True)
import warnings
warnings.filterwarnings("ignore")
```

```python
#1)PCA
df = pd.read_csv("/content/CC GENERAL.csv")
#Reading the CC General file
df.head()
```

```python
df.isnull().any()
# checking null data in the dataset using isnull() function
```

```python
df.fillna(df.mean(), inplace=True)
df.isnull().any()
#replacing the null data with the mean
```

```python
x = df.iloc[:,1:-1]
y = df.iloc[:,-1]
print(x.shape,y.shape)

```

```python
# a. Apply PCA on CC dataset.
#Datasets can be analyzed with PCA so that redundant features can be removed without losing too much information.
pca = PCA(3)   #Instantiate PCA
x_pca = pca.fit_transform(x)
principalDf = pd.DataFrame(data = x_pca, columns = ['principal component 1', 'principal component 2', 'principal component 3'])
finalDf = pd.concat([principalDf, df.iloc[:,-1]], axis = 1)
finalDf.head()
```

```python
# b. Apply k-means algorithm on the PCA result and report your observation if the silhouette score has improved or not?
X = finalDf.iloc[:,0:-1]
y = finalDf.iloc[:,-1]
print(X.shape,y.shape)
```

```python
nclusters = 3 # this is the k in kmeans
km = KMeans(n_clusters=nclusters)
km.fit(X)

# predict the cluster for each data point
y_cluster_kmeans = km.predict(X)

# Summary of the predictions made by the classifier
print(classification_report(y, y_cluster_kmeans, zero_division=1))
print(confusion_matrix(y, y_cluster_kmeans))

#finding the accuracy
train_accuracy = accuracy_score(y, y_cluster_kmeans)
print("\nAccuracy for our Training dataset with PCA:", train_accuracy)

#Calculating sihouette Score
score = metrics.silhouette_score(X, y_cluster_kmeans)
print("Sihouette Score: ",score)   #ranges from -1 to +1, high value shows that it is matched more
```

Question-4: Briefly identify the difference between PCA and LDA


Popular machine learning methods for reducing dimensionality include PCA (Principal Component Analysis) and LDA (Linear Discriminant Analysis). But their purposes and methods are different:

Purpose:
PCA is used for unsupervised learning and to identify a dataset's directions with the highest levels of variance. By converting the original dataset into a new coordinate system, where the features are uncorrelated and ranked by their variance, it lowers the number of features. Data compression, visualization, and noise reduction all frequently use PCA. Contrarily, LDA is employed for supervised learning and seeks to identify the linear feature combinations that most effectively distinguish the classes. By reducing the number of features and increasing class separability, the original dataset is projected onto a lower-dimensional space.LDA is commonly used for feature extraction, pattern recognition, and classification.

Method:
Finding the eigenvalues and eigenvectors of the data's covariance matrix is how PCA works. The eigenvalues show how much variance is explained by each eigenvector, and the eigenvectors show the directions of highest variance. PCA chooses the top k eigenvectors, where k is the desired reduced dataset's dimensionality.
In contrast, LDA maximizes the data's between-class scatter and reduces its within-class scatter. Finding the between-class scatter matrix and the within-class scatter matrix's product's eigenvectors and eigenvalues is required. The top k eigenvectors, which correspond to the biggest eigenvalues, are chosen via LDA.
